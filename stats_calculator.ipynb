```python
"""
=============================================================================
مشروع: الحاسبة الإحصائية الشاملة
(Statistical Calculator)
المطور:إيهاب المفلح
تاريخ التطوير: 21/07/2026
ملاحظات وإيضاحات حول المعادلة الإحصائية أثناء الحل
-----------------------------------------------------------------------------
- لحساب التباين
  (Variance)
  والانحراف المعياري
  (Standard Deviation):
  تم الاعتماد على تباين العينة
  (Sample Variance)
  المخصصة للعينات، والتي تقسم على
  (N - 1)
  بدلاً من
  N
  عند إدخال قيمة واحدة فقط
  ،(N = 1)
  تكون القسمة على
  ،(1 - 1 = 0)
  وهو ما يسبب خطأ القسمة على صفر
  (ZeroDivisionError)
  تم التعامل مع هذه الحالة برمجياً بجعل التباين والانحراف المعياري يساويان 0.0
  لضمان استقرار البرنامج وعدم توقفه فجأة
  (Crash)
=============================================================================
"""

import math

# ==============================================
# Step 1 : Building the main statistic formulas
# ==============================================

def calculate_count(numbers):
    """ترجع عدد العناصر بالقائمة"""
    return len(numbers)

def calculate_sum(numbers):
    """ترجع مجموع العناصر"""
    return sum(numbers)

def calculate_mean(numbers):
    """ترجع المتوسط الحسابي"""
    if not numbers:
        return 0.0
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    """ترجع الوسيط (مع مراعاة الأعداد الزوجية والفردية)"""
    if not numbers:
        return 0.0
    sorted_nums = sorted(numbers)
    n = len(sorted_nums)
    mid = n // 2

    if n % 2 == 0:
        return (sorted_nums[mid - 1] + sorted_nums[mid]) / 2.0
    else:
        return float(sorted_nums[mid])

def calculate_mode(numbers):
    """ترجع المنوال (القيمة أو القيم الأكثر تكراراً)"""
    if not numbers:
        return []

    # حساب تكرار كل عنصر باستخدام Dictionary
    counts = {}
    for num in numbers:
        counts[num] = counts.get(num, 0) + 1

    max_count = max(counts.values())

    # استخراج كل القيم التي تكررت بأعلى معدل
    modes = [num for num, count in counts.items() if count == max_count]
    return modes

def calculate_variance(numbers):
    """تحسب تباين العينة (Sample Variance)"""
    n = len(numbers)
    if n <= 1:
 # إذا كان العنصر واحداً أو القائمة فارغة تكون قيمة التباين 0 لعدم وجود تشتت
        return 0.0

    mean = calculate_mean(numbers)
    squared_diffs = sum((x - mean) ** 2 for x in numbers)
    return squared_diffs / (n - 1)

def calculate_std(numbers):
    """تحسب الانحراف المعياري (الجذر التربيعي للتباين)"""
    return math.sqrt(calculate_variance(numbers))

def find_min_max(numbers):
    """ترجع أصغر وأكبر قيمة كـ Tuple: (min, max)."""
    if not numbers:
        return (0.0, 0.0)
    return (min(numbers), max(numbers))

def calculate_range(numbers):
    """تحسب المدى (أكبر قيمة - أصغر قيمة)"""
    if not numbers:
        return 0.0
    min_val, max_val = find_min_max(numbers)
    return max_val - min_val


# ==============================================
# Step 2: Dealing with inputs and fixing errors
# ==============================================

def get_numbers_from_user():
    """
    تسمح للمستخدم بإدخال الأرقام مفصولة بفاصلة
    تعالج المسافات، المدخلات غير الرقمية، والقوائم الفارغة
    """
    while True:
        user_input = input("\nادخل الأرقام مفصولة بفاصلة (مثال: 15, 25, 43): ").strip()

        if not user_input:
            print("❌ خطأ: لم تدخل أي بيانات! يرجى إدخال رقم واحد على الأقل")
            continue

        parts = user_input.split(",")
        parsed_numbers = []
        has_error = False

        for part in parts:
            clean_part = part.strip()
            if clean_part == "":
                continue  # إهمال الفواصل الزائدة متتالية

            try:
                # تحويل النص إلى عدد عشري أو صحيح
                num = float(clean_part)
                # تحويل الأرقام الصحيحة مثل 12.0 إلى 12 لتحسين الشكل
                if num.is_integer():
                    num = int(num)
                parsed_numbers.append(num)
            except ValueError:
                print(f"❌ خطأ: القيمة '{clean_part}' ليست رقماً صالحاً! يرجى إعادة الإدخال بالكامل بشكل صحيح")
                has_error = True
                break

        if has_error:
            continue

        if not parsed_numbers:
            print("❌ خطأ: يرجى إدخال رقم واحد على الأقل")
            continue

        return parsed_numbers


# =============================================
# Step 3: Showing results in coordinated shape
# =============================================

def display_statistics(numbers):
    """تنادي كل الدوال الإحصائية وتطبع النتائج بشكل منظم وواضح"""
    if not numbers:
        print("⚠️ لا توجد بيانات لعرضها!")
        return

    count_val = calculate_count(numbers)
    sum_val = calculate_sum(numbers)
    mean_val = calculate_mean(numbers)
    median_val = calculate_median(numbers)
    modes = calculate_mode(numbers)
    var_val = calculate_variance(numbers)
    std_val = calculate_std(numbers)
    min_val, max_val = find_min_max(numbers)
    range_val = calculate_range(numbers)

    # تنسيق عرض المنوال في حال كانت قيمة واحدة أو قيم متعددة
    mode_str = ", ".join(str(m) for m in modes)

    print("\n=== نتائج التحليل الإحصائي ===")
    print(f"عدد القيم         : {count_val}")
    print(f"المجموع           : {sum_val}")
    print(f"المتوسط           : {mean_val:.2f}")
    print(f"الوسيط            : {median_val}")
    print(f"المنوال           : {mode_str}")
    print(f"التباين           : {var_val:.2f}")
    print(f"الانحراف المعياري : {std_val:.2f}")
    print(f"أصغر قيمة         : {min_val}")
    print(f"أكبر قيمة         : {max_val}")
    print(f"المدى             : {range_val}")
    print("===============================\n")


# ==========================================
# Step 4: Interactive Menu
# ==========================================

def main():
    last_analyzed_data = None  # متغير لتخزين آخر قائمة تم تحليلها

    while True:
        print("--- 📊 الحاسبة الإحصائية الشاملة 📊 ---")
        print("1. إدخال قائمة أرقام جديدة وتحليلها")
        print("2. عرض آخر نتيجة تحليل تمت")
        print("3. خروج من البرنامج")

        choice = input("اختر خياراً (1-3): ").strip()

        if choice == "1":
            numbers = get_numbers_from_user()
            last_analyzed_data = numbers
            display_statistics(numbers)

        elif choice == "2":
            if last_analyzed_data is None:
                print("\n⚠️ لا توجد نتائج سابقة محفوظة. قم بتحليل قائمة أولاً!\n")
            else:
                print("\n🔄 إعادة عرض آخر تحليل تم إجراؤه:")
                print(f"البيانات المُدخلة: {last_analyzed_data}")
                display_statistics(last_analyzed_data)

        elif choice == "3":
            print("\n👋 شكراً لاستخدامك الحاسبة الإحصائية الشاملة. إلى اللقاء!")
            break

        else:
            print("\n❌ خيار غير صحيح! يرجى إدخال 1 أو 2 أو 3\n")


# تشغيل البرنامج المباشر
if __name__ == "__main__":
    main()


# =============================================
# Step 5: Try the program on 3 different cases
# =============================================

# الحالة الأولى: إدخال قائمة أرقام عادية
# Inputs: 89, 23, 34, 67, 23, 45, 12

# الحالة الثانية: قائمة تحتوي على رقم واحد فقط
# Input: 42
# تم حساب التباين والانحراف المعياري بقيمة 0.00 بسلام بدون إعطاء خطأ القسمة على صفر

# الحالة الثالثة: محاولة إدخال قيمة نصية غير صحيحة أو مسافات بالخطأ
# Inputs: abc, 45, 12
# إعادة الطلب تلقائياً: يتزقف الكود عن الانهيار ويسأل المستخدم من جديد دون اغلاق البرنامج
```
