```python
import math

# 1. The Addition for 2 vectors formula:
def vector_add(v1, v2):
    # نجمع كل عنصرين متقابلين باستخدام zip
    return [x + y for x, y in zip(v1, v2)]

# Addition Manually (v1 + v2):
# [3+1,4+2] = [4,6]

# 2. دالة طرح متجهين
def vector_subtract(v1, v2):
    # نطرح كل عنصر من نظيره المتقابل
    return [x - y for x, y in zip(v1, v2)]

# Subtraction Manually (v1 - v2):
# [3-1,4-2] = [2,2]

# 3. دالة الضرب النقطي (Dot Product)
def dot_product(v1, v2):
    # نضرب كل عنصرين متقابلين ثم نجمع كافة النواتج
    return sum(x * y for x, y in zip(v1, v2))

# .Multiplication Manually sum (v1 * v2):
# (3*1) + (4*2) = (3) + (8) = 11

# 4. دالة حساب طول المتجه (Magnitude)
def magnitude(v):
    # الجذر التربيعي لمجموع مربعات العناصر
    return math.sqrt(sum(x ** 2 for x in v))

# Calculate magnitude v1 and v2 manually:
# the length direction for v1 (||v1||):
# √(3^2+4^2) = √(9+16) = √(25) = 5
# the length direction for v2 (||v2||):
# √(1^2+2^2) = √(1+4) = √(5) ≈ 2.2361

# 5. Testing and trying formulas with the given examples:
v1 = [3, 4]
v2 = [1, 2]

print("=== Results of running the codes ===")
print(f"v1 + v2          = {vector_add(v1, v2)}")
print(f"v1 - v2          = {vector_subtract(v1, v2)}")
print(f"Dot Product      = {dot_product(v1, v2)}")
print(f"Magnitude of v1  = {magnitude(v1)}")
print(f"Magnitude of v2  = {magnitude(v2):.4f}")
```

```python
# 1. The formula to sum two matrices in the same size:
def matrix_add(m1, m2):
    rows = len(m1)
    cols = len(m1[0])
    # بناء مصفوفة جديدة بحاصل جمع كل عنصرين متقابلين
    result = []
    for i in range(rows):
        row = []
        for j in range(cols):
            row.append(m1[i][j] + m2[i][j])
        result.append(row)
    return result


# 2. The formula to multiply two matrices (row * column):
def matrix_multiply(m1, m2):
    rows_m1 = len(m1)
    cols_m1 = len(m1[0])
    rows_m2 = len(m2)
    cols_m2 = len(m2[0])

    # Check the possibility of multiplication:
    if cols_m1 != rows_m2:
        raise ValueError("You can't multiply those matrices because the number of columns in matrix m1 must equal the number of rows in matrix m2!")

    # Creating the matrix for result in dimensions (rows_m1 x cols_m2) full of 0:
    result = [[0 for _ in range(cols_m2)] for _ in range(rows_m1)]

    # Nested for loops for matrix multplication
    for i in range(rows_m1):         # Walkthrough the rows in first matrix
        for j in range(cols_m2):     # Walkthrough the columns in second matrix
            for k in range(cols_m1): # Iterate through the common elements to calculate the row-by-column multiplication
                result[i][j] += m1[i][k] * m2[k][j]

    return result


# 3. Transpose Formula:
def transpose(m):
    rows = len(m)
    cols = len(m[0])

    # Changing rows into columns and columns into rows:
    result = []
    for j in range(cols):
        new_row = []
        for i in range(rows):
            new_row.append(m[i][j])
        result.append(new_row)
    return result


# 4. Trying and testing the multiplicatin of (2x2) matrix:
A = [
    [1, 2],
    [3, 4]
]

B = [
    [2, 0],
    [1, 3]
]

# Testing multiply two (2x2) matrices manually:
# element (1,1): (1x2) + (2x1) = 2 + 2 = 4
# element (1,2): (1x0) + (2x3) = 0 + 6 = 6
# element (2,1): (3x2) + (4x1) = 6 + 4 = 10
# element (2,2): (3x0) + (4x3) = 0 + 12 = 12
# A x b = ([[4,6] , [10,12]])

print("=== Results of running the codes ===")
print("A + B = ", matrix_add(A, B))
print("A x B = ", matrix_multiply(A, B))
print("Transpose(A) = ", transpose(A))
```

```python
import math

# 1. Mean Formula:
def calculate_mean(numbers):
    if not numbers:
        return 0
    return sum(numbers) / len(numbers)


# 2. Median Formula:
def calculate_median(numbers):
    if not numbers:
        return 0

    # Sorting numbers is important to calculate median:
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2

    # If the count of the numbers are even we will calculate the mean for the two numbers in the middle
    if n % 2 == 0:
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2.0
    else:
        # If the count of the numbers are odd we wiil take the number where is in the middle directly
        return float(sorted_numbers[mid])


# 3. Variance Formula:
def calculate_variance(numbers):
    if len(numbers) < 2:
        return 0.0

    mean = calculate_mean(numbers)
    # مجموع مربعات الفروقات عن المتوسط
    squared_diffs = sum((x - mean) ** 2 for x in numbers)

    # حساب تباين العينة (Sample Variance) بالقسمة على N - 1
    return squared_diffs / (len(numbers) - 1)


# 4. Standard Deviation Formula:
def calculate_std(numbers):
    # الانحراف المعياري هو الجذر التربيعي للتباين
    return math.sqrt(calculate_variance(numbers))


# 5. Testing formulas on grades list:
grades = [85, 90, 78, 92, 88, 76, 95]

mean_val = calculate_mean(grades)
median_val = calculate_median(grades)
var_val = calculate_variance(grades)
std_val = calculate_std(grades)

print("=== Statistics Results for Students Grades ===")
print(f"المتوسط الحسابي (Mean)        : {mean_val:.2f}")
print(f"الوسيط (Median)               : {median_val:.2f}")
print(f"التباين (Variance)            : {var_val:.2f}")
print(f"الانحراف المعياري (Std Dev)  : {std_val:.2f}")

# Explaining results:
# المتوسط الحسابي(86.29): يعبر عن المعدل العام لأداء جميع الطلاب في المادة.
# الوسيط (88.00): يمثل الدرجة التي تقع تماماً في منتصف ترتيب الطلاب (نصف الدرجات أعلى منها ونصفها أقل منها).
# التباين (52.24): يقيس متوسط مربع تشتت الدرجات وتباعدها عن المعدل العام.
# الانحراف المعياري (7.23): يوضح أن درجات أغلب الطلاب متقاربة نسبياً ومحتشدة ضمن نطاق 7.23 نقطة فقط تقريباً أعلى أو أقل من المتوسط العام (86.29).
```

```python
import math

# دوال مساعدة من الجزء السابق
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_std(numbers):
    mean = calculate_mean(numbers)
    # تباين العينة (قسمة على n - 1)
    variance = sum((x - mean) ** 2 for x in numbers) / (len(numbers) - 1)
    return math.sqrt(variance)

# 1. دالة حساب معامل ارتباط بيرسون يدوياً
def calculate_correlation(x, y):
    if len(x) != len(y) or len(x) < 2:
        raise ValueError("The two lists must have the same length and contain at least two elements!")

    n = len(x)
    mean_x = calculate_mean(x)
    mean_y = calculate_mean(y)

    std_x = calculate_std(x)
    std_y = calculate_std(y)

    # حساب التباين المشترك (Covariance)
    covariance = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n)) / (n - 1)

    # معامل الارتباط = التباين المشترك / (الانحراف المعياري لـ x * الانحراف المعياري لـ y)
    r = covariance / (std_x * std_y)
    return r

# 2. Testing formula on study hours and final grades data:
study_hours = [1, 2, 3, 4, 5, 6]
final_grades = [50, 55, 65, 70, 80, 85]

correlation_result = calculate_correlation(study_hours, final_grades)

print(f"معامل ارتباط بيرسون (r) = {correlation_result:.4f}")

# Explaining results:
# طبيعة العلاقة:العلاقة طردية قوية جداً، لأن قيمة معامل الارتباط r ≈ 0.995 وهي قريبة جدا من +1
# المعنى العملي: هذا يعني أنه كلما زادت ساعات المذاكرة، ترتفع الدرجة النهائية بشكل مباشر وشبه مؤكد.
```

```python
# 1. Probability of withdraw red ball:
red_balls = 5
blue_balls = 3
total_balls = red_balls + blue_balls

p_red = red_balls / total_balls

print(f"احتمال سحب كرة حمراء: {p_red:.4f} (أو {p_red * 100:.1f}%)")


# 2. احتمال الحصول على "صورة" مرتين متتاليتين عند رمي عملة (حدثين مستقلين)
# احتمال الصورة في الرمية الأولى = 0.5 وفي الرمية الثانية = 0.5
p_heads_1st = 1 / 2
p_heads_2nd = 1 / 2

# للحدثين المستقلين: P(A and B) = P(A) * P(B)
p_both_heads = p_heads_1st * p_heads_2nd

print(f"احتمال الحصول على صورة بالمرتين: {p_both_heads:.4f} (أو {p_both_heads * 100:.1f}%)")


# 3. الاحتمال الشرطي: P(نجاح | حضور)
# القانون: P(A|B) = عدد الحالات التي حققت الشرط والنتيجة معاً / عدد الحالات التي حققت الشرط
attended_students = 60          # الفئة المعتبرة (الشرط: الذين حضروا)
passed_and_attended = 54        # الذين نجحوا من نفس الفئة التي حضرت

p_pass_given_attendance = passed_and_attended / attended_students

print(f"احتمال نجاح الطالب بشرط أنه حضر كل المحاضرات: {p_pass_given_attendance:.4f} (أو {p_pass_given_attendance * 100:.1f}%)")
```
