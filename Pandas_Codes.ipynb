```python
# Pandas is a shortcut for panel data
# Typically works with objects known as Series and DataFrame: a series is like a one dimensional labled column, a dataframe is like a two dimensional table
# in this library you can import, Display, Manipulate, Export tagular data
# Think of panda as python version of microsoft Excel

import pandas as pd
print(pd.__version__) # to check the version of pandas library
```

```python
import pandas as pd
# Series = A Pandas 1-Dimensional labled array that can hold any data type. Think of it like a single column in spreadsheet (1-Dimensional)
data = [100, 102, 104]
series = pd.Series(data)
print(series)

data = [100.1, 102.2, 104.3]  # to change the data type to float
series = pd.Series(data)
print(series)

data = ["A", "B", "C"]  # to change the data type to object (string)
series = pd.Series(data)
print(series)

data = [True, False, True]  # to change the data type boolean
series = pd.Series(data)
print(series)

data = [100, 102, 104]
series = pd.Series(data, index=["a", "b", "c"]) # to change the indexes to custom labels you can use any type of arrays
print(series)

data = [100, 102, 104]
series = pd.Series(data, index=["apartment #1", "apartment #2", "apartment #3"]) # to change the indexes to custom labels you can use any type of arrays
print(series)
```

```python
import pandas as pd
data = [100, 102, 104]
series = pd.Series(data, index=["a", "b", "c"])
# you can return a value of a label:
print(series.loc["a"])
print(series.loc["b"])
print(series.loc["c"])
# if you put print(series,loc["d"]) you will get value error because label d is not exist

data = [100, 102, 104]
series = pd.Series(data, index=["a", "b", "c"])
series.loc["c"] = 200 # to change the value of the label(index)
print(series.loc["c"])
print(series)

data = [100, 102, 104]
series = pd.Series(data, index=["a", "b", "c"])
print(series.iloc[0]) # to locate the value by integer position we use iloc
print(series.iloc[1]) # to locate the value by integer position we use iloc
print(series.iloc[2]) # to locate the value by integer position we use iloc
```

```python
import pandas as pd
data = [100, 102, 104, 200, 202]
series = pd.Series(data, index=["a", "b", "c", "d", "e"]) # labeles (indexes) must equal the number of values
print(series [series >= 200]) # To return any value greater than or equal to 200
print(series [series < 200]) # To return any value lower than 200
```

```python
import pandas as pd
# Complex Example:
# I like pizza I need to go on  a diet because am eating too much pizza am going to use a python dictionary to record how many calories I've eaten per day
# I'll try to eat less than 2000 calories each day

calories = {"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}
series = pd.Series(calories) # since we have a key values no need to pass an index like before in the list because each value is already labeled
print(series)

print(series.loc["Day 1"])
print(series.loc["Day 2"])
print(series.loc["Day 3"])

# lets update Day 3 perhaps am cheated on my diet:
calories = {"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}
series = pd.Series(calories)
series.loc["Day 3"] += 500 # we added 500 calories
print(series.loc["Day 3"])
print(series)

# Now filter by value: lets find any days on which I've eaten more than 2000 calories:
calories = {"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}
series = pd.Series(calories)
print(series[series >= 2000]) # on this day I didn't stick to my diet
print(series[series < 2000]) # on those days I did stick to my diet
```

```python
# DataFrame = A tabular data structure with rows AND columns. (2 Dimensional) similar to an Excel spreadsheet
import pandas as pd

data = {
    "Name": ["Spongebob", "Patrick", "Squidward"],
    "Age": [30, 35, 50]
}
df = pd.DataFrame(data) # we used the name df as a shortcut for dataframe
print(df)

# to change the indexes to be labeled:
data = {
    "Name": ["Spongebob", "Patrick", "Squidward"],
    "Age": [30, 35, 50]
}
df = pd.DataFrame(data, index= ["Employee 1", "Employee 2", "Employee 3"])
print(df)
print(df.loc["Employee 1"]) # to take a data from selected row, it will give me the name and the age of Employee 1 with the label name (index) and data type
print(df.loc["Employee 2"]) # to take a data from selected row, it will give me the name and the age of Employee 2 with the label name (index) and data type
print(df.loc["Employee 3"]) # to take a data from selected row, it will give me the name and the age of Employee 3 with the label name (index) and data type

# you can also select by integer location:
print(df.iloc[0])
print(df.iloc[1])
print(df.iloc[2])
```

```python
import pandas as pd
# If you want to add a new column & row:
data = {
    "Name": ["Spongebob", "Patrick", "Squidward"],
    "Age": [30, 35, 50]
}
df = pd.DataFrame(data, index= ["Employee 1", "Employee 2", "Employee 3"])
# Add a new Column:
df["Job"] = ["Cook", "N/A", "Cashier"] # The number of values in column must equal the number of values in rows
df["Salary"] = [3000, 2000, 1000]
# add a new row the best way is to create a new dataframe then to use concatenate:
new_row = pd.DataFrame([{"Name": "Sandy", "Age": 28, "Job": "Engineer", "Salary": 700}], index=["Employee 4"]) # New list with new dictonary inside it
df = pd.concat([df, new_row]) # to merge the original dataframe with the new one
print(df)
# to cahnge the last index instead of 0 to become Employee 4 I should add a new index called "Employee 4" in the new dataframe
```

```python
import pandas as pd
# to add multi rows
data = {
    "Name": ["Spongebob", "Patrick", "Squidward"],
    "Age": [30, 35, 50]
}
df = pd.DataFrame(data, index= ["Employee 1", "Employee 2", "Employee 3"])
# Add a new Column:
df["Job"] = ["Cook", "N/A", "Cashier"]
df["Salary"] = [3000, 2000, 1000]
# add a new rows the best way is to create a new dataframe then to use concatenate:
new_rows = pd.DataFrame([{"Name": "Sandy", "Age": 28, "Job": "Engineer", "Salary": 700},
                         {"Name": "Eugene", "Age": 60, "Job": "Manager", "Salary": 7000}], index=["Employee 4", "Employee 5"]) # New list with new dictionaries inside it
df = pd.concat([df, new_rows]) # to merge the original dataframe with the new one
print(df)
```

```python
import pandas as pd
df = pd.read_csv("/data.csv")
# Filtering = Keeping the rows that match a condition
tall_pokemon = df[df["Height"] >= 2] # to return any pokemon its tall is greater than or equal to 2, here we created a new datafrmae
heavy_pokemon = df[df["Weight"] > 100] # to return any pokemon its weight is greater than 100, here we created another new datafrmae
legendary_pokemon = df[df["Legendary"] == 1] # to return any legendary pokemon its value is 1 if not legendary its value is 0
# you could write it also like this: legendary_pokemon = df[df["Legendary"] == True] because the boolean value of True = 1
water_pokemon = df[df["Type1"] == "Water"] # to return any pokemon its Type1 is Water
print(tall_pokemon)
print(heavy_pokemon)
print(legendary_pokemon)
print(water_pokemon)

water_pokemon = df[(df["Type1"] == "Water") | (df["Type2"] == "Water")] # The symbol | means or, to return any pokemon has either in Type1 or Type2 Water
print(water_pokemon)

ff_pokemon = df[df["Type1"] == "Fire"] # to return any pokemon its Type1 is Fire
print(ff_pokemon)
ff_pokemon = df[(df["Type1"] == "Fire") & (df["Type2"] == "Flying")] # to return any pokemon its Type1 is Fire and Type2 is Flying
# make sure to not use python and insted of &
print(ff_pokemon)
```

```python
# aggregate functions = Reduces a set of values into a single summary value used to summaraize and analyze data Often used with groupby() function
# Those aggregate functions will be applied to whole dataframe
import pandas as pd
df = pd.read_csv("/data.csv")
print (df.mean(numeric_only=True)) # because you can't calculate the mean for non-numeric columns like "Name"
print (df.sum(numeric_only=True)) # to calculate the sum of the values in each numeric column
print (df.min(numeric_only=True)) # to calculate the minimum value in each numeric column
print (df.max(numeric_only=True)) # to return the maximum value from each numeric column
print (df.count()) # to count the number of values in each column, Type2 is 67 because NaN will not count
```

```python
# We can arrange our dataframe into diifrent groups
# For example we can put all the Grass into one group and all the Fire types into another one group and do that for all the different types
# We need to use groupby function
import pandas as pd
df = pd.read_csv("/data.csv")
group = df.groupby("Type1")
print(group) # if you put it ;ike this it will give you its an object
print(group["Height"].mean()) # to get the average height for each group in Type1
print(group["Height"].sum()) # to get the sum of height in each group in Type1
print(group["Height"].min()) # to get the minimum value in each group in Type1
print(group["Height"].max()) # to get the maximum value in each group in Type1
print(group["Height"].count()) # to count the number of values in each group in Type1
```

```python
# Data Cleaning: the process of fixing/removing:
#                incomplete, incorrect, or irrelevant data.
#                ~75% of work donewith pandas is data cleaning
import pandas as pd
df = pd.read_csv("/data.csv")
 # to drop for example an irrelevant column, we will drop "Legendary Column" here, you can drop more columns by addinf them to the list separate by commas
df = df.drop(columns=["Legendary"]) # we created a new dataframe to not drop the column from the original dataframe
print(df) # now "Legendary" column is missing
```

```python
# Handle Missing Data
# in my data I have pokemons have Type1 but dont have type2 which is NaN so we need to drop the rows that contains NaN
import pandas as pd
df = pd.read_csv("/data.csv")
df = df.dropna(subset=["Type2"]) # dropna is a shortcut for not available, here in column "Type2" drop any rows that missing a value
print(df.to_string())
```

```python
# Handle Missing Data
import pandas as pd
df = pd.read_csv("/data.csv")
df = df.fillna({"Type2": "None"}) # fillna is a shortcut for fill any not available values, we will use a dictionary,
# here it means replace any not available value within column Type2 with the follwing value instead of Not a Number (NaN) we will replace these values with None
print(df.to_string())
```

```python
# Fix inconsistent values
import pandas as pd
df = pd.read_csv("/data.csv")
df["Type1"] = df["Type1"].replace({"Grass": "GRASS"}) # to change the whole Grass in column "Type1" to become GRASS
print (df.to_string())
```
