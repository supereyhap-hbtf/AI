```python
import math

# ==========================================
# 1. Tools for (Vectors)
# ==========================================

def vector_add(v1, v2):
    """ترجع ناتج جمع متجهين عنصرين عنصرين"""
    return [x + y for x, y in zip(v1, v2)]

def vector_subtract(v1, v2):
    """ترجع ناتج طرح متجهين عنصرين عنصرين"""
    return [x - y for x, y in zip(v1, v2)]

def dot_product(v1, v2):
    """تحسب الضرب النقطي (مجموع حاصل ضرب العناصر المتقابلة)"""
    return sum(x * y for x, y in zip(v1, v2))

def magnitude(v):
    """تحسب طول المتجه (الجذر التربيعي لمجموع مربعات عناصره)"""
    return math.sqrt(sum(x ** 2 for x in v))


# ==========================================
# 2. Tools for (Matrices)
# ==========================================

def matrix_add(m1, m2):
    """تجمع مصفوفتين بنفس الأبعاد"""
    return [[m1[i][j] + m2[i][j] for j in range(len(m1[0]))] for i in range(len(m1))]

def matrix_multiply(m1, m2):
    """تضرب مصفوفتين بالتأكد من تساوي أعمدة الأولى مع صفوف الثانية"""
    rows_m1, cols_m1 = len(m1), len(m1[0])
    cols_m2 = len(m2[0])
    result = [[0 for _ in range(cols_m2)] for _ in range(rows_m1)]
    for i in range(rows_m1):
        for j in range(cols_m2):
            for k in range(cols_m1):
                result[i][j] += m1[i][k] * m2[k][j]
    return result

def transpose(m):
    """تقلب صفوف المصفوفة إلى أعمدة والأعمدة إلى صفوف"""
    return [[m[i][j] for i in range(len(m))] for j in range(len(m[0]))]


# ==========================================
# 3. Tools for (Statistics)
# ==========================================

def calculate_mean(numbers):
    """تحسب المتوسط الحسابي لقائمة أرقام"""
    return sum(numbers) / len(numbers) if numbers else 0.0

def calculate_median(numbers):
    """تحسب الوسيط لقائمة أرقام"""
    if not numbers:
        return 0.0
    sorted_nums = sorted(numbers)
    n = len(sorted_nums)
    mid = n // 2
    return (sorted_nums[mid - 1] + sorted_nums[mid]) / 2.0 if n % 2 == 0 else float(sorted_nums[mid])

def calculate_variance(numbers):
    """تحسب تباين العينة لقائمة أرقام"""
    if len(numbers) < 2:
        return 0.0
    mean = calculate_mean(numbers)
    return sum((x - mean) ** 2 for x in numbers) / (len(numbers) - 1)

def calculate_std(numbers):
    """تحسب الانحراف المعياري لقائمة أرقام"""
    return math.sqrt(calculate_variance(numbers))


# ==========================================
# 4. Tools for (Correlation & Probability)
# ==========================================

def calculate_correlation(x, y):
    """تحسب معامل ارتباط بيرسون بين متغيرين"""
    n = len(x)
    mean_x, mean_y = calculate_mean(x), calculate_mean(y)
    std_x, std_y = calculate_std(x), calculate_std(y)
    covariance = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n)) / (n - 1)
    return covariance / (std_x * std_y)

def calculate_probability(success_count, total_count):
    """تحسب احتمال حدث بسيط (الحالات الناجحة / الحالات الكلية)"""
    return success_count / total_count if total_count > 0 else 0.0


# ==========================================
# Practical Application on Students Grades
# ==========================================

if __name__ == "__main__":
    # 2. إنشاء بيانات تجريبية لـ 10 طلاب
    students_data = [
        {"name": "أحمد", "study_hours": 2, "grade": 55},
        {"name": "سارة", "study_hours": 8, "grade": 92},
        {"name": "خالد", "study_hours": 4, "grade": 68},
        {"name": "منى", "study_hours": 6, "grade": 82},
        {"name": "عمر", "study_hours": 1, "grade": 45},
        {"name": "فاطمة", "study_hours": 9, "grade": 95},
        {"name": "يوسف", "study_hours": 3, "grade": 60},
        {"name": "ريم", "study_hours": 5, "grade": 75},
        {"name": "علي", "study_hours": 7, "grade": 88},
        {"name": "هند", "study_hours": 2, "grade": 50},
    ]

    # استخراج القوائم كمتجهات
    study_hours = [student["study_hours"] for student in students_data]
    grades = [student["grade"] for student in students_data]

    # 3. الحسابات باستخدام المكتبة
    mean_grade = calculate_mean(grades)
    std_grade = calculate_std(grades)
    correlation_val = calculate_correlation(study_hours, grades)

    # حساب احتمال النجاح (الدرجة >= 60)
    passing_students = sum(1 for g in grades if g >= 60)
    pass_probability = calculate_probability(passing_students, len(grades))

    # عرض النتائج
    print("=== نتائج تحليل أداء الطلاب ===")
    print(f"• متوسط درجات الطلاب      : {mean_grade:.2f}")
    print(f"• الانحراف المعياري للدرجات : {std_grade:.2f}")
    print(f"• معامل الارتباط (M - G)  : {correlation_val:.4f}")
    print(f"• احتمال نجاح طالب عشوائي   : {pass_probability * 100:.1f}%")

    # ملخص نتائج أداء الطلاب
    # يُظهر التحليل الإحصائي للبيانات أن أداء الطلاب ممتاز بشكل عام، حيث بلغ متوسط الدرجات حوالي 71 درجة من 100، مع انحراف معياري قدره 18.09
    # يعكس هذا الانحراف وجود تفاوت ملحوظ بين أداء الطلاب؛ إذ نجد درجات مرتفعة جداً تقارب 95وأخرى منخفضة في حدود 45
    # أما على صعيد فرص النجاح، فإن احتمالية اختيار طالب ناجح عشوائياً من هذه المجموعة هي 70% (حيث نجح 7 طلاب من أصل 10)
    # أما بالنسبة للعامل المؤثر في هذا الأداء، فقد أظهر تحليل معامل الارتباط رقماً مرتفعاً جداً يبلغ 0.98 (قريب جداً من +1)
    # يُترجم هذا الرقم عملياً إلى وجود علاقة طردية قوية استثنائياً بين ساعات المذاكرة والدرجة النهائية
    # مما يعني بعبارة بسيطة: كلما زاد الطالب عدد ساعات دراسته الأسبوعية، ارتفعت درجته النهائية بشكل شبه مضمون ومباشر
```
