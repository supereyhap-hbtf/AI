```python
# دالة التحقق من الرقم الأولي is_prime
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

# --- أسطر تشغيل واختبار الدالة ---
num_to_test = 17
result = is_prime(num_to_test)
print(f"هل الرقم {num_to_test} أولي؟ {result}")

num_to_test_2 = 4
print(f"هل الرقم {num_to_test_2} أولي؟ {is_prime(num_to_test_2)}")
```

```python
# دالة حساب المساحة calculate_area(shape, ...)
def calculate_area(shape, length=None, width=None, radius=None):
    shape = shape.lower()
    if shape in ["rectangle", "مستطيل"]:
        if length is not None and width is not None:
            return length * width
        return "خطأ: يجب تمرير الطول والعرض."
    elif shape in ["circle", "دائرة"]:
        if radius is not None:
            return 3.14159 * (radius ** 2)
        return "خطأ: يجب تمرير نصف القطر."
    return "الشكل غير مدعوم."

# --- أسطر تشغيل واختبار الدالة ---
# اختبار حساب مساحة مستطيل (طوله 5 وعرضه 4)
rect_area = calculate_area("مستطيل", length=5, width=4)
print(f"مساحة المستطيل: {rect_area}")

# اختبار حساب مساحة دائرة (نصف قطرها 3)
circle_area = calculate_area("دائرة", radius=3)
print(f"مساحة الدائرة: {circle_area:.2f}")
```

```python
# دالة عكس النص reverse_string
def reverse_string(text):
    reversed_text = ""
    for char in text:
        reversed_text = char + reversed_text
    return reversed_text

# --- أسطر تشغيل واختبار الدالة ---
my_text = "Python"
final_result = reverse_string(my_text)
print(f"النص الأصلي: {my_text} -> النص المعكوس: {final_result}")
```

```python
# دالة حساب المتوسط average
def average(numbers_list):
    if len(numbers_list) == 0:
        return 0
    total_sum = sum(numbers_list)
    count = len(numbers_list)
    return total_sum / count

# --- أسطر تشغيل واختبار الدالة ---
my_list = [10, 20, 30, 40, 50]
list_average = average(my_list)
print(f"متوسط الأرقام في القائمة {my_list} هو: {list_average}")
```

```python
# نظام إدارة المخزون المتكامل

# 1. هيكل البيانات الرئيسي (Dictionary) لتخزين المنتجات
inventory = {
    "laptop": {"quantity": 10, "price": 500.0, "category": "electronics"},
    "mouse": {"quantity": 50, "price": 15.0, "category": "electronics"}
}

# 2. مجموعة (Set) لتخزين فئات المنتجات الفريدة دون تكرار
categories = {"electronics"}


def display_categories():
    """دالة مساعدة لعرض الفئات الحالية المتوفرة في المخزن باستخدام الـ Set"""
    print(f"📁 الفئات المتوفرة حالياً في المخزن: {', '.join(categories)}")


def add_product():
    """1. إضافة منتج جديد تماماً للمخزن"""
    print("\n--- إضافة منتج جديد ---")
    name = input("أدخل اسم المنتج الجديد: ").strip().lower()

    if name in inventory:
        print(f"❌ المنتج '{name}' موجود بالفعل! لتعديل كميته استخدم خيار التعديل من القائمة الرئيسية.")
        return

    try:
        quantity = int(input("أدخل الكمية الابتدائية: "))
        price = float(input("أدخل سعر المنتج: "))
        if quantity < 0 or price < 0:
            print("❌ خطأ: لا يمكن إدخال كمية أو سعر بقيم سالبة!")
            return
    except ValueError:
        print("❌ خطأ: يرجى إدخال قيم رقمية صالحة للكمية والسعر!")
        return

    category = input("أدخل فئة المنتج (مثال: electronics, office): ").strip().lower()
    if not category:
        category = "general"  # فئة افتراضية في حال لم يدخل المستخدم شيئاً

    # إضافة المنتج للمخزن الرئيسي
    inventory[name] = {
        "quantity": quantity,
        "price": price,
        "category": category
    }

    # إضافة الفئة إلى الـ Set لضمان تسجيلها كفئة فريدة
    categories.add(category)

    print(f"✅ تم إضافة المنتج الجديد '{name}' بنجاح ضمن فئة '{category}'!")


def update_quantity():
    """2. تعديل كمية منتج موجود (زيادة عند الشراء / نقصان عند البيع)"""
    print("\n--- تعديل كمية منتج ---")
    name = input("أدخل اسم المنتج المراد تعديل كميته: ").strip().lower()

    if name not in inventory:
        print("❌ خطأ: هذا المنتج غير موجود في المخزن!")
        return

    current_qty = inventory[name]["quantity"]
    print(f"الكمية الحالية المتوفرة من '{name}' هي: {current_qty}")
    print("اختر نوع العملية:")
    print("1. زيادة الكمية (شراء بضاعة جديدة)")
    print("2. نقصان الكمية (عملية بيع)")

    action = input("اختر (1 أو 2): ").strip()

    if action not in ["1", "2"]:
        print("❌ خيار غير صحيح!")
        return

    try:
        amount = int(input("أدخل الكمية المراد تعديلها: "))
        if amount <= 0:
            print("❌ يجب أن تكون الكمية المراد تعديلها أكبر من الصفر!")
            return
    except ValueError:
        print("❌ خطأ: يرجى إدخال رقم صحيح للكمية!")
        return

    if action == "1":
        # عملية شراء وزيادة
        inventory[name]["quantity"] += amount
        print(f"➕ تم إضافة {amount} وحدة. الكمية الجديدة لـ '{name}' هي: {inventory[name]['quantity']}")

    elif action == "2":
        # عملية بيع ونقصان مع معالجة تخطي الحد المتاح لمنع الـ Crash أو القيم السالبة
        if amount > current_qty:
            print(f"❌ عملية مرفوضة! الكمية المطلوبة للبيع ({amount}) أكبر من الكمية المتوفرة بالمخزن ({current_qty}).")
        else:
            inventory[name]["quantity"] -= amount
            print(f"➖ تم بيع {amount} وحدة. الكمية المتبقية من '{name}' هي: {inventory[name]['quantity']}")


def remove_product():
    """3. حذف منتج بالكامل من المخزن"""
    print("\n--- حذف منتج بالكامل ---")
    name = input("أدخل اسم المنتج المراد حذفه: ").strip().lower()

    if name in inventory:
        # حذف المنتج
        del inventory[name]
        print(f"🗑️ تم حذف المنتج '{name}' بالكامل من المخزن.")

        # إعادة تحديث الـ Set الخاص بالفئات بناءً على العناصر المتبقية فعلياً في المخزن
        global categories
        categories = {details["category"] for details in inventory.values()}
    else:
        print("❌ خطأ: هذا المنتج غير موجود بالأساس في المخزن!")


def display_inventory():
    """4. عرض كل المنتجات بجدول منسّق"""
    print("\n" + "="*55)
    print(f"{'المنتج':<15} | {'الفئة':<15} | {'الكمية':<8} | {'السعر ($)':<10}")
    print("="*55)

    if not inventory:
        print("المخزن فارغ حالياً!")
    else:
        for product, details in inventory.items():
            print(f"{product:<15} | {details['category']:<15} | {details['quantity']:<8} | {details['price']:<10.2f}")

    print("="*55)
    display_categories()  # استدعاء دالة عرض الفئات الفريدة (Set) أسفل الجدول
    print()


def search_product():
    """5. البحث عن منتج معين بالاسم"""
    print("\n--- البحث عن منتج ---")
    name = input("أدخل اسم المنتج للبحث عنه: ").strip().lower()

    if name in inventory:
        details = inventory[name]
        print("\n🔍 تم العثور على المنتج:")
        print(f"• الاسم: {name}")
        print(f"• الفئة: {details['category']}")
        print(f"• الكمية المتوفرة: {details['quantity']}")
        print(f"• سعر الوحدة: {details['price']:.2f} $")
        print(f"• القيمة الإجمالية لهذا المنتج: {details['quantity'] * details['price']:.2f} $\n")
    else:
        print("❌ لم يتم العثور على المنتج المطلوب في المخزن.")


def calculate_total_value():
    """6. حساب القيمة المالية الإجمالية لكل المخزن"""
    total_value = 0.0
    for details in inventory.values():
        total_value += details["quantity"] * details["price"]
    print(f"\n💰 القيمة المالية الإجمالية للمخزون بالكامل هي: {total_value:.2f} $\n")


def main():
    """الحلقة التكرارية الرئيسية لتشغيل نظام القائمة التفاعلية"""
    while True:
        print("=== نظام إدارة المخزون ===")
        print("1. إضافة منتج")
        print("2. تعديل كمية")
        print("3. حذف منتج")
        print("4. عرض كل المنتجات")
        print("5. البحث عن منتج")
        print("6. القيمة الإجمالية للمخزون")
        print("7. خروج")

        choice = input("اختر رقم العملية (1-7): ").strip()

        if choice == "1":
            add_product()
        elif choice == "2":
            update_quantity()
        elif choice == "3":
            remove_product()
        elif choice == "4":
            display_inventory()
        elif choice == "5":
            search_product()
        elif choice == "6":
            calculate_total_value()
        elif choice == "7":
            print("\n👋 تم إغلاق النظام بنجاح. شكراً لك!")
            break
        else:
            print("❌ خيار غير صحيح! يرجى إدخال رقم من 1 إلى 7 وإعادة المحاولة.\n")


# تشغيل الكود البرمجي فوراً عند تشغيل الملف
if __name__ == "__main__":
    main()
```
